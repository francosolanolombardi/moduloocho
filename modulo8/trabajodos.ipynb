```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Abrir base y depurarla

```python
df_csv = pd.read_csv("owid-covid-data.csv")
columnas_interes = ['iso_code','continent','location','date','total_cases','new_cases','total_deaths','new_deaths','total_cases_per_million','new_cases_per_million',
                    'total_deaths_per_million','new_deaths_per_million','total_tests','total_tests_per_thousand','total_vaccinations','people_vaccinated','people_fully_vaccinated','people_fully_vaccinated_per_hundred','new_vaccinations',
                    'people_vaccinated_per_hundred','population_density','median_age','aged_70_older','extreme_poverty','population','cardiovasc_death_rate',
                    'diabetes_prevalence','human_development_index'
                    ]
df_prueba = df_csv[columnas_interes]
df = df_prueba.dropna()
df
df.to_csv("base_final_trabajo.csv")
```

### Distribuci√≥n

```python
# Variables categ√≥ricas
#Cantidad de registros por continente
plt.figure(figsize=(8,5))
ax = sns.countplot(data=df, x="continent", order=df["continent"].value_counts().index, palette="Blues")

# A√±adir el valor encima de cada barra
for p in ax.patches:
    altura = p.get_height()
    ax.text(p.get_x() + p.get_width()/2, altura + 5, 
            int(altura), 
            ha="center")

plt.title("Cantidad de filas por continente")
plt.xlabel("Continente")
plt.ylabel("Cantidad de filas")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# Variables categ√≥ricas
# Cantidad de filas por mes
# Crear una columna con fecha tipo periodo mensual para ordenar y graficar
df['date'] = pd.to_datetime(df['date'])

# Extraer a√±o y mes en columnas nuevas
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month

# Agrupar por a√±o y mes y contar filas
conteo_mensual = df.groupby(['year', 'month']).size().reset_index(name='cantidad_filas')
conteo_mensual['periodo'] = pd.to_datetime(conteo_mensual['year'].astype(str) + '-' + conteo_mensual['month'].astype(str) + '-01')

conteo_mensual = conteo_mensual.sort_values('periodo')

plt.figure(figsize=(12,6))
plt.plot(conteo_mensual['periodo'], conteo_mensual['cantidad_filas'], marker='o')
plt.title('Cantidad de registros por mes y a√±o')
plt.xlabel('Fecha')
plt.ylabel('Cantidad de registros')
plt.grid(False)
plt.show()
```

```python
columnas_numericas = df.select_dtypes(include="number").columns

for col in columnas_numericas:
    plt.figure(figsize=(6, 4))
    sns.histplot(data=df, x=col, kde=True)
    plt.title(f"Distribuci√≥n de {col}")
    plt.tight_layout()
    plt.show()
```

### Outliers

```python
#Outliers
columnas_numericas = df.select_dtypes(include="number").columns

resumen_outliers = []

for col in columnas_numericas:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    outliers = df[(df[col] < limite_inferior) | (df[col] > limite_superior)]
    cantidad_outliers = len(outliers)

    resumen_outliers.append({
        "Variable": col,
        "Q1": Q1,
        "Q3": Q3,
        "IQR": IQR,
        "L√≠mite Inferior": limite_inferior,
        "L√≠mite Superior": limite_superior,
        "Cantidad de Outliers": cantidad_outliers
    })

tabla_outliers = pd.DataFrame(resumen_outliers)

tabla_outliers[["Variable","Cantidad de Outliers"]]
```

```python
#Box plot por continente
columnas_numericas = df.select_dtypes(include="number").columns

for col in columnas_numericas:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df, x="continent", y=col)
    plt.title(f"Distribuci√≥n de {col} por continente")
    plt.xlabel("Continente")
    plt.ylabel(col)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

### Relaci√≥n entre variables

```python

# Seleccionar solo columnas num√©ricas
columnas_numericas = df.select_dtypes(include="number")

# Calcular la matriz de correlaci√≥n
matriz_correlacion = columnas_numericas.corr()

plt.figure(figsize=(15, 15))
sns.heatmap(matriz_correlacion, annot=True, fmt=".2f", cmap="coolwarm", square=True, cbar=False)
plt.title("Matriz de correlaci√≥n")
plt.tight_layout()
plt.show()
```
